### 3. Identify the predominant age group among employees.

```python
myData['Age'].min()
```

```python
myData['Age'].max()
```

```python
employees
```

```python
bins = [0 , 20 , 30, 40]
labels = ["0-20", "21-30", "31-40"]

# Create age groups using cut()
myData["Age_Group"] = pd.cut(myData["Age"], bins=bins, labels=labels, right=False)

# Count employees in each age group
age_group_counts = myData["Age_Group"].value_counts()

# Find the most predominant age group
predominant_group = age_group_counts.idxmax()

print("Age Group Counts:")
print(age_group_counts)
print("\nMost Predominant Age Group:", predominant_group)
```

### Graphical Representation:

```python
sns.histplot(myData["Age_Group"],color='r')
```

### 4.Discover which team and position have the highest salary expenditure. 

```python
salary_expenditure = myData.groupby(['Team','Position'])['Salary'].sum()
x,y=salary_expenditure.idxmax()
print("Team with Highest Salary Expenditure is - ", x)
print("Position with Highest Salary Expenditure is - ",y)
```

### 5.Investigate if there's any correlation between age and salary, and represent it visually.
